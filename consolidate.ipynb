```python
import pandas as pd 
import json 
import os 
```

```python
relationship_files = [i for i in os.walk("./relationships")][0][2]
with open("./person.json") as infile:
    actor_dict = json.load(infile)
    
```

```python
relationship_dict = {
    "id": [],
    "from_actor": [],
    "to_actor": [],
    "start": [],
    "end": [],
    "role": [],
    "from_actor_name": [],
    "to_actor_name": [],
    "type": []
}
```

```python
for i in relationship_files: 
    print(i)

    with open(f"./relationships/{i}") as infile: 
        tmp_dict = json.load(infile)
    actor = list(tmp_dict.keys())[0]
    try: 
        relationship_list = [j for j in tmp_dict[actor]]
    except KeyError:
        continue 
    for j in relationship_list: 
        if int(actor) == j['to_actor']: 
            if j['start'] == None and j['end'] == None:
                continue 
            else: 
                relationship_dict['from_actor'].append(j['to_actor'])
                relationship_dict['from_actor_name'].append(j['to_actor_name'])
                relationship_dict['to_actor'].append(j['from_actor'])
                relationship_dict['to_actor_name'].append(j['from_actor_name'])
                relationship_dict['id'].append(j['id'])
                relationship_dict['start'].append(j['start'])
                relationship_dict['end'].append(j['end'])
                relationship_dict['role'].append(j['role'])
                relationship_dict['type'].append(j['type'])
        else: 
            if j['start'] == None and j['end'] == None:
                continue 
            else: 
                for key in relationship_dict: 
                    if key in ['start', 'end']:
                        relationship_dict[key].append(str(j[key]))
                    else: 
                        relationship_dict[key].append(j[key])
```

```python
relationship_df = pd.DataFrame(relationship_dict)
relationship_df.to_csv("./relationships.csv", sep=";",index=False)
```
